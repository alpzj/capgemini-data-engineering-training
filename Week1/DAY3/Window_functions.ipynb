### 1.	Assign a unique row number to all employees based on salary (highest first). 

```sql
select *,
row_number() over(order by salary DESC) as row_number
from employees
```

### 2.	Assign row numbers to employees within each department based on salary descending. 

```sql
select *,row_number() over(partition by department order by salary desc) as dept_row_num from employees
```

### 3.	Assign row numbers based on employee joining date (latest first). 

```sql
select *,
row_number() over(order by join_date desc) as join_row_number
from employees
```

### 4.	Assign row numbers within each department based on earliest joining date. 

```sql
select *,
row_number() over(partition by department order by join_date asc) as department_join_rank
from employees
```
