### Loading the dataset

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Spark DataFrames").getOrCreate()
df = spark.read.csv("/Volumes/workspace/default/datasets/C1.csv",header="True")
df.show()
```

### Displaying Data

```python
df.display()
```

```python
df.printSchema()
```

### Replacing "blank" and "Blank" with NULL

```python
df.na.replace(["blank","Blank"],None).display()
```

### First Letter Capital In Country Column

```python
from pyspark.sql.functions import *
df=df.withColumn("Country",when(col("Country")==upper(col("Country")),col("Country")).otherwise(initcap(col("Country"))))
df.display()
```

### Change of date format 

```python
 df.withColumn("JoinDate",coalesce(try_to_date(col("JoinDate"),"dd-MM-yyyy"),try_to_date(col("JoinDate"),"dd-MM-yyyy"))).display()
```

### Replace Country

```python
df.replace("New York","USA").display()
```

### Remove Duplicates

```python
df.dropDuplicates().display()
```

```python
df=df.filter(col("CustomerID")!=104)
df=df.orderBy("CustomerID")
df.display()
```
