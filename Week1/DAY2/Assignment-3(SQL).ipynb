### 6. Find the total salary for departments where the total salary exceeds 100,000.

### 16. Find the total sales amount for each employee

```sql
select e.emp_id,e.emp_name,sum(s.amount) as total_sales from Employee e
left join sales s on s.emp_id=e.emp_id
group by e.emp_id,e.emp_name
```

### 17. List the number of sales made by each employee.

```sql
select e.emp_id,e.emp_name,count(s.amount) as sales_count from Employee e
left join sales s on s.emp_id=e.emp_id
group by e.emp_id,e.emp_name
```

### 18. Find the total sales amount grouped by product.

```sql
SELECT product, SUM(amount) AS total_sales
FROM sales
GROUP BY product;
```

### 19. Calculate the average sales amount grouped by product.

```sql
SELECT product, AVG(amount) AS total_sales
FROM sales
GROUP BY product;
```

### 20. Find employees who have made more than 2 sales, grouped by their names.

```sql
select e.emp_id,e.emp_name,count(s.sales_id) as sales_count
from Employee e
join sales s on s.emp_id=e.emp_id
group by e.emp_id,e.emp_name
having count(s.sales_id)>2
```

### 21. Calculate the total salary and total sales amount for each employee.

```sql
select e.emp_id,e.emp_name,e.salary,coalesce(sum(s.amount),0) as total_sales
from employee e 
left join sales s on s.emp_id=e.emp_id
group by e.emp_id,e.emp_name,e.salary
```

### 22. Count the number of unique products sold by each employee.

```sql
SELECT e.emp_id,e.emp_name,COUNT(DISTINCT s.product) AS unique_products
FROM Employee e
JOIN Sales s ON e.emp_id = s.emp_id
GROUP BY e.emp_id, e.emp_name
```

### 23. Find the highest sales amount made by each employee.

```sql
select e.emp_name,e.emp_id,max(s.amount) as highest_sale
from employee e 
join sales s on s.emp_id=e.emp_id
group by e.emp_id,e.emp_name
order by highest_sale desc
```

### 24. Calculate the total sales amount grouped by product and filtered by products where the total exceeds 50,000.


```sql
select product,sum(amount) as total_sales
from sales
group by product
having total_sales>50000
```

### 25. Find the departments with the highest average sales amount.

```sql
select e.department,avg(s.amount) as avg_amount
from Employee e join sales s on e.emp_id=s.emp_id
group by e.department
order by avg_amount desc
```

### 26. Find the department with the highest total sales amount.

```sql
SELECT e.department, SUM(s.amount) AS total_sales
FROM Employee e
JOIN Sales s ON e.emp_id = s.emp_id
GROUP BY e.department
ORDER BY total_sales DESC
LIMIT 1
```

### 27. Show the top 3 employees with the highest total sales amount, grouped by employee names.


```sql
select e.emp_name,e.emp_id,sum(s.amount) as total_sales from employee e
join sales s on s.emp_id=e.emp_id
group by e.emp_name,e.emp_id
order by total_sales desc
limit 3
```

### 28. Calculate the total number of employees and the average salary, grouped by the year of joining.


```sql
select year(joining_date) as year1,count(emp_id) as total_employees,avg(salary) as avg_salary
from employee
group by year1
```

### 29. Find the total sales amount for each department (using a join between Employee and Sales).



```sql
select e.department,sum(s.amount) as total_sales from sales s
join employee e on s.emp_id=e.emp_id
group by e.department
```

### 30. Show employees who have not made any sales, grouped by their department

```sql
SELECT  e.emp_id,e.emp_name,e.department
FROM Employee e
LEFT JOIN Sales s ON e.emp_id=s.emp_id
WHERE s.sales_id IS NULL
GROUP BY e.emp_id,e.emp_name,e.department;
```
