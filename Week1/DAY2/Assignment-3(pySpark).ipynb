### Importing required libraries

```python
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

### Creating Employee DataFrame

```python
employee_data = [
    (1, 'Karthik', 'HR', 60000.00, '2021-01-15'),
    (2, 'Pratik', 'Finance', 70000.00, '2021-03-10'),
    (3, 'Veer', 'HR', 55000.00, '2021-06-20'),
    (4, 'Priya', 'Finance', 80000.00, '2022-01-05'),
    (5, 'Ajay', 'Engineering', 75000.00, '2020-11-01'),
    (6, 'Vijay', 'Engineering', 78000.00, '2019-05-22'),
    (7, 'Veena', 'HR', 62000.00, '2023-03-12'),
    (8, 'Meena', 'Marketing', 65000.00, '2022-08-18')
]

employee_schema = StructType([
    StructField("emp_id", IntegerType(), True),
    StructField("emp_name", StringType(), True),
    StructField("department", StringType(), True),
    StructField("salary", DoubleType(), True),
    StructField("joining_date", StringType(), True)
])

employee_df = spark.createDataFrame(employee_data, schema=employee_schema)
```

### Convert joining_date to DateType

```python
from pyspark.sql.functions import to_date, col

employee_df = employee_df.withColumn(
    "joining_date",
    to_date(col("joining_date"), "yyyy-MM-dd")
)
```

### Create Sales DataFrame

```python
sales_data = [
    (1, 1, 'Laptop', 50000.00, '2023-01-15'),
    (2, 2, 'Mobile', 30000.00, '2023-02-18'),
    (3, 3, 'Tablet', 20000.00, '2023-02-25'),
    (4, 4, 'Laptop', 45000.00, '2023-03-05'),
    (5, 5, 'Mobile', 35000.00, '2023-03-12'),
    (6, 6, 'Tablet', 25000.00, '2023-03-20'),
    (7, 7, 'Laptop', 60000.00, '2023-04-01'),
    (8, 8, 'Mobile', 40000.00, '2023-04-10')
]

sales_schema = StructType([
    StructField("sales_id", IntegerType(), True),
    StructField("emp_id", IntegerType(), True),
    StructField("product", StringType(), True),
    StructField("amount", DoubleType(), True),
    StructField("sale_date", StringType(), True)
])

sales_df = spark.createDataFrame(sales_data, schema=sales_schema)
```

### Convert sale_date to DateType

```python
sales_df = sales_df.withColumn(
    "sale_date",
    to_date(col("sale_date"), "yyyy-MM-dd")
)
```

### 1. Find the total salary for each department in the Employee table.

```python
employee_df.groupBy("department").agg(sum("salary").alias("Total_Salary")).display()
```

### 2. Count the number of employees in each department.

```python
employee_df.groupBy("department").agg(count("emp_id").alias("Total_Employees")).display()
```

### 3. Calculate the average salary of employees in each department.

```python
employee_df.groupBy("department").agg(avg("salary").alias("Average_Salary")).display()
```

### 4. Find the maximum salary in each department.

```python
employee_df.groupBy("department").agg(max("salary").alias("Max_Salary")).display()
```

### 5. Find the minimum salary in each department.


```python
employee_df.groupBy("department").agg(min("salary").alias("Min_Salary")).display()
```

### 6. Find the total salary for departments where the total salary exceeds 100,000.

```python
employee_df.groupBy("department").agg(sum("salary").alias("Total_Salary")).filter(col("Total_Salary")>100000).display()
```

```python
employee_df.groupBy("department").agg(count("emp_id").alias("Total_Employees")).filter(col("Total_Employees")>2).display()
```
