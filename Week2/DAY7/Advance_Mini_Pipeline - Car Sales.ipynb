### Phase 2 – Cleaning

### 1. Trim strings

```python
customers = customers.withColumn("name", trim(col("name"))) \
                     .withColumn("city", trim(col("city"))).display()

dealers = dealers.withColumn("name", trim(col("name"))) \
                 .withColumn("city", trim(col("city"))).display()
```

### 2. Fix negative prices

```python
cars = cars.withColumn("price", when(col("price") < 0, abs(col("price")))
                                     .otherwise(col("price"))).display()
```

### 3. Remove duplicates

```python
customers = customers.dropDuplicates()
cars = cars.dropDuplicates()
sales = sales.dropDuplicates()
```

### Phase 3 – Validation

### Find invalid foreign keys

```python
# Invalid customer_id in sales
invalid_customers = sales.join(customers, "customer_id", "left_anti")

# Invalid car_id
invalid_cars = sales.join(cars, "car_id", "left_anti")

# Invalid dealer mapping
invalid_dealers = sales_dealer.join(dealers, "dealer_id", "left_anti")
```

### Remove invalid rows

```python
sales_valid = sales.join(customers, "customer_id") \
                   .join(cars, "car_id")

sales_dealer_valid = sales_dealer.join(dealers, "dealer_id")
```

### Phase 4 – Transformation

### Build Base Fact Table (FIX for your broken pipeline)

```python
df = sales_valid \
    .join(sales_dealer_valid, "sale_id") \
    .join(dealers, "dealer_id") \
    .withColumn("revenue", col("price") * col("quantity"))
```

### 1. Customer Revenue

```python
customer_revenue = df.groupBy("customer_id") \
                     .agg(sum("revenue").alias("total_revenue"))
```

### 2. Brand-wise Sales

```python
brand_sales = df.groupBy("brand") \
                .agg(sum("revenue").alias("brand_revenue"))
```

### 3. City-wise Revenue

```python
city_revenue = df.join(customers, "customer_id") \
                 .groupBy("city") \
                 .agg(sum("revenue").alias("city_revenue"))
```

### Phase 5 – Dealer Analytics

### 1. Revenue per Dealer

```python
dealer_revenue = df.groupBy("dealer_id") \
                   .agg(sum("revenue").alias("dealer_revenue"))
```

### 2. Top Dealers

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number, desc

w = Window.orderBy(desc("dealer_revenue"))

top_dealers = dealer_revenue.withColumn("rank", row_number().over(w)) \
                            .filter(col("rank") <= 5)
```

### 3. Dealer City Contribution

```python
dealer_city = df.groupBy("dealer_id", "city") \
                .agg(sum("revenue").alias("revenue"))
```

### Phase 6 – SQL Analysis

### Register temp views

```python
df.createOrReplaceTempView("sales_fact")
customers.createOrReplaceTempView("customers")
```

### 1. Top 3 customers per city

```python
%sql
SELECT *
FROM (
    SELECT c.city,
           s.customer_id,
           SUM(s.revenue) AS revenue,
           ROW_NUMBER() OVER (PARTITION BY c.city ORDER BY SUM(s.revenue) DESC) rn
    FROM sales_fact s
    JOIN customers c ON s.customer_id = c.customer_id
    GROUP BY c.city, s.customer_id
) t
WHERE rn <= 3;
```

### 2. Monthly Trends

```python
%sql
SELECT date_format(sale_date, 'yyyy-MM') AS month,
       SUM(revenue) AS total_revenue
FROM sales_fact
GROUP BY month
ORDER BY month;
```

### 3. Repeat Customers

```python
%sql
SELECT customer_id, COUNT(*) AS purchases
FROM sales_fact
GROUP BY customer_id
HAVING COUNT(*) > 1;
```
