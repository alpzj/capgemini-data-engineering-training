### Phase 2 – Data Cleaning 

### Handle Null Values

```python
from pyspark.sql.functions import col
customers_clean = customers_df.dropna(subset=["customer_id"])
policies_clean = policies_df.dropna(subset=["policy_id", "customer_id"])
claims_clean = claims_df.dropna(subset=["claim_id", "policy_id"])
agents_clean = agents_df.dropna(subset=["agent_id"])

customers_clean = customers_clean.fillna({"name": "Unknown","city": "Unknown","age": 0})
policies_clean = policies_clean.fillna({"policy_type": "Unknown","premium": 0})
claims_clean = claims_clean.fillna({"claim_amount": 0,"status": "Unknown"})
agents_clean = agents_clean.fillna({"name": "Unknown","region": "Unknown"})
```

### Handle Negative Values

```python
from pyspark.sql.functions import when
policies_clean = policies_clean.withColumn("premium",when(col("premium") < 0, 0).otherwise(col("premium")))
claims_clean = claims_clean.withColumn("claim_amount",when(col("claim_amount") < 0, 0).otherwise(col("claim_amount")))
```

### Standardize String Columns

```python
from pyspark.sql.functions import trim, lower, initcap

customers_clean = customers_clean.withColumn("name", trim(col("name")))
customers_clean = customers_clean.withColumn("city", initcap(trim(col("city"))))
policies_clean = policies_clean.withColumn("policy_type", initcap(trim(col("policy_type"))))
claims_clean = claims_clean.withColumn("status", initcap(trim(col("status"))))
agents_clean = agents_clean.withColumn("name", trim(col("name")))
agents_clean = agents_clean.withColumn("region", initcap(trim(col("region"))))
```

### Ensure Correct Data Types

```python
from pyspark.sql.functions import to_date

policies_clean = policies_clean.withColumn(
    "start_date", to_date(col("start_date"))
)

claims_clean = claims_clean.withColumn(
    "claim_date", to_date(col("claim_date"))
)
```

### Remove Duplicates

```python
customers_clean = customers_clean.dropDuplicates(["customer_id"])
policies_clean = policies_clean.dropDuplicates(["policy_id"])
claims_clean = claims_clean.dropDuplicates(["claim_id"])
agents_clean = agents_clean.dropDuplicates(["agent_id"])
```

### Create Final Clean Dataset

```python
final_clean_df = customers_clean \
    .join(policies_clean, "customer_id", "left") \
    .join(claims_clean, "policy_id", "left") \
    .join(policy_df, "policy_id", "left") \
    .join(agents_clean, "agent_id", "left")

final_clean_df.display(5)
```

### Phase 3 – Data Validation 

### Invalid Foreign Key Checks

```python
## Policies with NO matching Customer
invalid_policies = policies_clean.join(customers_clean,on="customer_id",how="left_anti")
## Claims with NO matching Policy
invalid_claims = claims_clean.join(policies_clean,on="policy_id",how="left_anti")
## Policy-Agent with invalid Policy
invalid_policy_agent = policy_df.join(policies_clean,on="policy_id",how="left_anti")
## Policy-Agent with invalid Agent
invalid_agents = policy_df.join(agents_clean,on="agent_id",how="left_anti")

invalid_agents.show()
invalid_policy_agent.show()
invalid_claims.show()
invalid_policies.show()
```

### Count Invalid Records

```python
invalid_policies_count = invalid_policies.count()
invalid_claims_count = invalid_claims.count()
invalid_policy_agent_count = invalid_policy_agent.count()
invalid_agents_count = invalid_agents.count()

print("Invalid Policies:", invalid_policies_count)
print("Invalid Claims:", invalid_claims_count)
print("Invalid Policy-Agent (policy):", invalid_policy_agent_count)
print("Invalid Policy-Agent (agent):", invalid_agents_count)
```

### Total Records Count

```python
total_customers = customers_clean.count()
total_policies = policies_clean.count()
total_claims = claims_clean.count()
total_agents = agents_clean.count()
total_policy_agent = policy_df.count()

print("Total Customers:", total_customers)
print("Total Policies:", total_policies)
print("Total Claims:", total_claims)
print("Total Agents:", total_agents)
print("Total Policy-Agent:", total_policy_agent)
```

### Cleaned Records

```python
valid_policies = total_policies - invalid_policies_count
valid_claims = total_claims - invalid_claims_count
valid_policy_agent = total_policy_agent - invalid_policy_agent_count
```

### Phase 4 – Transformation 

### Total Premium per Customer

```python
from pyspark.sql.functions import sum
premium_per_customer = policies_clean.groupBy("customer_id").agg(sum("premium").alias("total_premium"))
premium_per_customer.display()
```

### Total Claim per Customer

```python
claim_per_customer = policies_clean \
    .join(claims_clean, "policy_id", "left") \
    .groupBy("customer_id") \
    .agg(sum("claim_amount").alias("total_claim"))

claim_per_customer.show()
```

### Combine Both (Customer Level Metrics)

```python
customer_metrics = premium_per_customer \
    .join(claim_per_customer, "customer_id", "left")

customer_metrics.show()
```

### Risk Score Calculation

```python
from pyspark.sql.functions import when, col

customer_metrics = customer_metrics.withColumn(
    "risk_score",
    when(col("total_premium") == 0, 0)
    .otherwise(col("total_claim") / col("total_premium"))
)

customer_metrics.show()
```

### Add Customer Info

```python
customer_final = customer_metrics.join(
    customers_clean.select("customer_id", "name", "city"),
    "customer_id",
    "left"
)

customer_final.show()
```

### Combine City Metrics

```python
%sql
-- Customer Risk Calculation using CTE
WITH premium_cte AS (
    SELECT 
        customer_id,
        SUM(premium) AS total_premium
    FROM policies
    GROUP BY customer_id
),
claims_cte AS (
    SELECT 
        p.customer_id,
        SUM(c.claim_amount) AS total_claim
    FROM policies p
    LEFT JOIN claims c 
        ON p.policy_id = c.policy_id
    GROUP BY p.customer_id
),
risk_cte AS (
    SELECT 
        p.customer_id,
        p.total_premium,
        COALESCE(c.total_claim, 0) AS total_claim,
        CASE 
            WHEN p.total_premium = 0 THEN 0
            ELSE c.total_claim / p.total_premium
        END AS risk_score
    FROM premium_cte p
    LEFT JOIN claims_cte c
        ON p.customer_id = c.customer_id
)
SELECT * FROM risk_cte;
```

### Top 3 Risky Customers per City

```python
%sql
WITH premium_cte AS (
    SELECT customer_id, SUM(premium) AS total_premium
    FROM policies
    GROUP BY customer_id
),
claims_cte AS (
    SELECT p.customer_id, SUM(c.claim_amount) AS total_claim
    FROM policies p
    LEFT JOIN claims c 
        ON p.policy_id = c.policy_id
    GROUP BY p.customer_id
),
risk_cte AS (
    SELECT 
        p.customer_id,
        p.total_premium,
        COALESCE(c.total_claim, 0) AS total_claim,
        CASE 
            WHEN p.total_premium = 0 THEN 0
            ELSE c.total_claim / p.total_premium
        END AS risk_score
    FROM premium_cte p
    LEFT JOIN claims_cte c
        ON p.customer_id = c.customer_id
),
rank_cte AS (
    SELECT 
        r.*,
        c.city,
        ROW_NUMBER() OVER (PARTITION BY c.city ORDER BY r.risk_score DESC) AS rank
    FROM risk_cte r
    JOIN customers c
        ON r.customer_id = c.customer_id
)
SELECT *
FROM rank_cte
WHERE rank <= 3;
```

### Customers with Increasing Claims

```python
%sql 
WITH monthly_claims AS (
    SELECT 
        p.customer_id,
        DATE_TRUNC('month', c.claim_date) AS month,
        SUM(c.claim_amount) AS monthly_claim
    FROM policies p
    JOIN claims c
        ON p.policy_id = c.policy_id
    GROUP BY p.customer_id, DATE_TRUNC('month', c.claim_date)
),
lag_cte AS (
    SELECT 
        customer_id,
        month,
        monthly_claim,
        LAG(monthly_claim) OVER (
            PARTITION BY customer_id 
            ORDER BY month
        ) AS prev_month_claim
    FROM monthly_claims
),
increase_flag AS (
    SELECT *,
        CASE 
            WHEN monthly_claim > prev_month_claim THEN 1
            ELSE 0
        END AS is_increasing
    FROM lag_cte
)
SELECT *
FROM increase_flag
WHERE is_increasing = 1;
```
